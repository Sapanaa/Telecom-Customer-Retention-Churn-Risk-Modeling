## Business Context

Telecommunication companies face customer attrition (churn), which directly impacts revenue and growth.

Acquiring new customers is significantly more expensive than retaining existing ones.

This project simulates a telecom provider aiming to:
- Identify high-risk customers
- Prioritize retention campaigns
- Reduce revenue loss due to churn


## Problem Statement

Given historical customer data, build a predictive model to estimate churn probability.

The model should:
- Accurately identify customers likely to churn
- Support targeted retention strategies
- Minimize revenue loss


## Business Cost Considerations

False Negative (Missed Churner):
- Customer leaves
- Revenue lost
- High replacement cost

False Positive (Wrongly Flagged Customer):
- Marketing cost (discount/offer)
- Lower financial impact

Therefore:
Recall is more important than overall accuracy.


## Evaluation Metrics

Primary Metric:
- Recall (for churn class)

Secondary Metrics:
- ROC-AUC
- Precision
- F1-score

Reason:
Missing a churner is more costly than contacting a loyal customer.


## Business Objective

The goal is to:
- Identify top 20% highest-risk customers
- Estimate potential revenue at risk
- Propose targeted retention strategies


## Expected Deliverables

- SQL-based churn KPI analysis
- Exploratory data analysis
- Predictive model comparison
- Threshold optimization
- Retention strategy simulation
